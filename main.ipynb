```python
import matplotlib.pyplot as plt
from cv2 import imread
import numpy as np
```

```python
R = imread('R.jpg') # idx2
G = imread('G.jpg') # idx1 
B = imread('B.jpg') # idx0 
```

```python
R.shape
```

```python
red_c = R[:,:,2] 
plt.imshow(red_c, cmap='Reds')
```

```python
green_c = G[:,:,1]
plt.imshow(green_c, cmap='Greens')
```

```python
blue_c = B[:,:,0]
plt.imshow(blue_c, cmap='Blues')
```

```python
combined = np.array([
    red_c,
    green_c,
    blue_c
])
```

```python
combined = combined.swapaxes(0,2).swapaxes(0,1)
```
